#Алгоритм Бена-Исраэля

```python
import numpy as np

eps = 0.05
point = np.array([-5,2,1,7])
A = np.array([[7,0,4,9,6,0],[3,0,4,1,1,2],[0,1,1,2,6,7],[1,4,7,4,9,5]])

def print_matrix(A):
  for row in A:
    for i in row:
      #print("{} & ".format(str(round(i))), end = "" )
      print("{}\t\t".format(str(round(i))), end = "")
    print()

def print_float_matrix(A):
  for row in A:
    for i in row:
      #print("{:6f} & ".format(i), end = "" )
      print("{:6f}\t\t".format(i), end = "")
    print()

def Ben_Israel(A, eps):
  I = np.diag(np.diag(np.ones([len(A),len(A)])))
  buf = 1
  i = 0

  if(np.linalg.norm(A).any()): #если не нулевая норма
    A1 = (A.T) / np.linalg.norm(A)
    A2 = A1.dot(2*I) - (A1.dot(A)).dot(A1)
    #E = np.linalg.norm(A2) - np.linalg.norm(A1) #||A + B|| <= ||A|| + ||B|| - свойство
    E = np.linalg.norm(A2 - A1) #разные размерности у матриц ОШИБКА

  else: #если нулевая норма
    A1 = np.zeros((len(A[0]), len(A)))
    buf = 0

  while(E >= eps and buf):
    A1 = A2.copy()
    A2 = A1.dot(2*I) - (A1.dot(A)).dot(A1)
    #E = np.linalg.norm(A2) - np.linalg.norm(A1) 
    E = np.linalg.norm(A2 - A1)

    print("\nШаг ", i, ":")
    print("A^+:")
    print_matrix(A1)
    print()
    i += 1


  return A1

Ben_Israel(A, eps)
```
